```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
request=[]
def scrape_page(url):
    driver.get(url)
    # Wait for the elements to be present
    wait = WebDriverWait(driver, 10)

    # Wait for the requested text elements to be present
    requested_text_elements = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.text-center small.ellipsis')))

    # Wait for the media heading elements to be present
    media_heading_elements = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.media-heading.ellipsis')))

    # Extract and print the requested text and media heading
    for requested_text, media_heading in zip(requested_text_elements, media_heading_elements):
        print("Requested Text:", requested_text.text)
        print("Media Heading:", media_heading.text)
        request.append(media_heading.text)
# Start scraping from page 1 to page 50
driver = web_driver()
for page_number in range(1, 100):
    page_url = f'https://simbi.com/requests?page={page_number}'
    scrape_page(page_url)

# Close the browser window
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920,1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
service=[]
def scrape_page(url):
    driver.get(url)
    # Wait for the elements to be present
    wait = WebDriverWait(driver, 10)

    # Wait for the card title elements to be present
    card_title_elements = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.card-title h4.ellipsis')))

    # Extract and print the card titles
    for card_title in card_title_elements:
        print("Card Title:", card_title.text)
        service.append(card_title.text)
# Start scraping from page 1 to page 50
driver = web_driver()
for page_number in range(1, 100):
    page_url = f'https://simbi.com/services/search?page={page_number}'
    scrape_page(page_url)

# Close the browser window
driver.quit()
```

```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained Sentence-BERT model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')



# Get embeddings for both lists
embeddings1 = model.encode(request, convert_to_tensor=True)
embeddings2 = model.encode(service, convert_to_tensor=True)

# Find most similar pairs based on cosine similarity
most_similar_pairs = []
for i, embed1 in enumerate(embeddings1):
    similarities = util.pytorch_cos_sim(embed1, embeddings2)[0]
    max_similarity_idx = similarities.argmax()
    most_similar_pair = (request[i], service[max_similarity_idx], similarities[max_similarity_idx].item())
    most_similar_pairs.append(most_similar_pair)

# Print most similar pairs
print("Most similar pairs are:")
for pair in most_similar_pairs:
    print(pair)
```
